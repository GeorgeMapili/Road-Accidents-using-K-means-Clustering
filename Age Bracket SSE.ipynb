```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel("Datasets/MetroManilaAccidentsAgeBracket.xlsx")
df.head()
```

```python
df.pivot_table(index="Year", columns="Age Bracket", aggfunc="sum", values="Total", margins=True)
```

```python
data=df.iloc[:,0:2].values
```

```python
print(data)
```

# Cluster in each datapoints

```python
cluster = KMeans(init="random", n_clusters = 3, n_init = 10, max_iter=300, random_state=42)
model = cluster.fit(data)
sse = cluster.inertia_
clusters = model.labels_
print(clusters)
```

# (Sum of Squared Error)SSE

```python
# calculated (Sum of Squared Error)SSE console 
#between SSE / SSE Calculation  
print(sse)
```

# (Sum of Squared Regression)SSR

```python
ssr = np.sum((cluster.inertia_ - data.mean())**2)
print(ssr)
```

# (Sum of Squared Total)SST

```python
sst = sse / ssr
print(sst)
```
